```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from skimage import data
import math
from skimage.color import rgb2hsv
from skimage.color import rgb2gray
from skimage.filters import threshold_otsu
```

```python
img1 = mpimg.imread('football.jpg')
```

```python
fig1=plt.figure()

subplot1=fig1.add_subplot(2,2,1)
plt.imshow(img1)
subplot1.set_title('Original Image')

img2 = rgb2hsv(img1)

subplot2=fig1.add_subplot(2,2,2)
plt.imshow(img2)
subplot2.set_title('HSV Image')

img3=rgb2gray(img1)

subplot3=fig1.add_subplot(2,2,3)
plt.imshow(img3, cmap="gray")
subplot3.set_title('Grayscale Image')

thresh = threshold_otsu(img3)
img4 = img3 > thresh
img4 = img4.astype(int)

subplot4=fig1.add_subplot(2,2,4)
plt.imshow(img4, cmap="gray")
subplot4.set_title('Binary Image')

plt.savefig('fig1')
```
